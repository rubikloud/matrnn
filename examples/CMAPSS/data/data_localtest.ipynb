````python
def get_single(groupedby, intraining, verbose = 0):
    
    unitdex, subset = groupedby
    maxtimetemp = np.array(subset).shape[0]
    startat = MAXT - maxtimetemp
    
    subset = np.array(subset)
    subset[np.isnan(subset)] = 0

    # 4 = #(tse,tte,unc,pcs)
    datshape_single = (1, MAXT, 4+NCOV)

    # set up output
    out = np.empty(datshape_single)
    out[:] = nanis
    
    # load covariates data 
    out[0, startat:, 4:4+NCOV] = subset[:, 2:]

    
    # time series stuff
    # failure occured at final time step
    indicator = np.zeros(maxtimetemp+1)
    indicator[-1] = 1
    if verbose:
        print ('indicator:\n', indicator)

    # tse, tte
    tse, tte = tstf.tse(indicator), tstf.tte(indicator)
    if verbose:
        print ('tse:\n', tse)
        print ('tte:\n', tte)
        
    out[0, startat:, 0] = tse[:maxtimetemp]
    out[0, startat:, 1] = tte[:maxtimetemp]

    # unc
    # all observations in training are uncensored if intraining
    # arrival times are censored if otherwise (no impact)
    if intraining:
        out[0, startat:, 2] = 1
    else:
        out[0, startat:, 2] = 0
    #out[0, -1, 2] = 0

    # pcs
    # if pcs==0: loglike is 0
    out[0, startat:, 3] = 1

    if verbose:
        print ('testing single t=0:\n', out[0, 0, :])
        print ('testing single t=1:\n', out[0, 1, :])
        print ('testing single t=startat-1:\n', out[0, startat-1, :])
        print ('testing single t=startat:\n', out[0, startat, :])
        print ('testing single t=startat+1:\n', out[0, startat+1, :])
        print ('testing single t=-2:\n', out[0, -2, :])
        print ('testing single t=-1:\n', out[0, -1, :])
        
    return out
````

````python
get_single(list(d.groupby('unitdex'))[0], intraining = False, verbose = 1)
````

````python
mtest = np.concatenate(list(map(lambda x: get_single(x, intraining=False), d.groupby('unitdex'))))
````
